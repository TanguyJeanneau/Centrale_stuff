### 1.1

Some beautiful graphs with various options: Axis, grid, title, vertical lines, text, plot, etc.

```python
T = 10
X_triangle = np.linspace(-3, 3, 7)
Y_triangle = (-1)**X_triangle
plt.axis(ymax=1.2, ymin=-1.2, xmin=-3*T/2, xmax=3*T/2)
plt.grid(color='gray', linestyle='--')
plt.xticks([0])
plt.yticks([-1, 0, 1])
plt.title('triangle')
plt.ylabel('$x(t)$', size=15)
plt.vlines([-T/2, T/2], ymin=-1, ymax=0, linestyles='--', color='gray')
plt.text(-T/1.5, 0.1, '$-T/2$')
plt.text(T/2.5, 0.1, '$T/2$')
plt.plot(T/2*X_triangle, Y_triangle)
```

```python
X_rectangle = np.linspace(-3, 2.999, 1000)
Y_rectangle = (-1)**np.floor(X_rectangle)
plt.grid(color='gray', linestyle='--')
plt.axis(ymax=1.2, ymin=-1.2, xmin=-3, xmax=3)
plt.xticks([0])
plt.yticks([-1, 0, 1])
plt.title('rectangle')
plt.ylabel('$x(t)$', size=15)
plt.plot(X_rectangle, Y_rectangle)
```

### 1.2 
What about the fourier transform of these two signals ? Here we compute the first 14 ones for triangle and square signals

```python
coefs_triangle = [8 / (np.pi*(2*k+1)) ** 2 for k in range(1, 15)]
coefs_rectangle = [4 / (np.pi*(2*k+1)) for k in range(1, 15)]
print(coefs_triangle)
print(coefs_rectangle)
```

### 1.3 & 1.4

Now let's plot the spectral coefs of both signals altogether with a nice subplot

```python
x_stem = range(1, 15)

plt.subplot(2, 1, 2)
plt.stem(x_stem, coefs_rectangle, use_line_collection=True)
plt.title('$rectangle$')
plt.ylabel('$b_{k}$', size=15)
plt.axis(xmin=0, xmax=14, ymin=0, ymax=0.5)
plt.yticks([0, 0.5])

plt.subplot(2, 1, 1)
plt.stem(x_stem, coefs_triangle, use_line_collection=True)
plt.title('$triangle$')
plt.ylabel('$a_{k}$', size=15)
plt.axis(xmin=0, xmax=14, ymin=0, ymax=0.1)
plt.yticks([0, 0.1])

plt.tight_layout()
plt.savefig('ex1_3.png')
```

### 1.5
Last but not least, we can use the computed coefs to display an approximation of a square signal with different $n$

```python
x = np.linspace(-np.pi, np.pi, 100)

def signal(n, t):
    l = [4/(np.pi*(2*k+1)) * np.sin((2*k+1)*np.pi*t) for k in range(0, n)]
    return sum(l)

for i in range(6):
    n = 2*i+1
    y = [signal(n, t) for t in x]
    plt.subplot(3, 2, i+1)
    plt.plot(x, y)
    plt.title('$n = {}$'.format(n))
    plt.xticks([])
    plt.yticks([])
    
plt.tight_layout()
```

## Exercice 2
Let's play around with some statistics

### 2.1
First of all, we'll build some basic functions and apply them to our examples $X$ and $Y$

```python
X = [190, 128, 180, 212, 56, 192, 68, 98, 110, 197, 181, 233]
Y = [24, 28, 24, 19, 37, 22, 34, 25, 36, 24, 20, 18]
```

```python
def mean(l):
    return sum(l)/len(l)

def median(l):
    tmp = l.copy()
    tmp.sort()
    return tmp[np.int(len(l)/2)]

def var(l):
    L = np.array(l)
    return ((L - mean(L)) ** 2).sum()/len(L)

def std(l):
    return np.sqrt(var(l))
```

```python
print('--------X----------Y--------')
print('median: {0:.3e}  {1:.3e}'.format(median(X), median(Y)))
print('mean:   {0:.3e}  {1:.3e}'.format(mean(X), mean(Y)))
print('var:    {0:.3e}  {1:.3e}'.format(var(X), var(Y)))
print('std:    {0:.3e}  {1:.3e}'.format(std(X), std(Y)))
```

### 2.2
What would be a statistics exercise without a good old linear regression ?

```python
(a, b) = np.polyfit(X, Y, deg=1)
print('Y = {0:.3e} X + {1:.3e}'.format(a, b))
```

### 2.3
Plotting time !

```python
plt.scatter(X, Y, marker='d')
plt.plot([0, 250], [b, 250*a+b],
         linestyle='--', color='red')
plt.xlabel('$X$')
plt.ylabel('$Y$')
```
