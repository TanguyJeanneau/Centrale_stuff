# Kalaman filter :

## Estimating a constant :

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def Kalman(N, R=1, Q=1e-5, C=1):
    #Initialization :
    Lz=[]
    Lx_hat=[]
    Lt=[]
    x_hat=-2
    P=1
    x=np.random.normal(x_hat,P)
    t=0
    while t<N:
        P=P+Q
        t+=1
        x=C+np.random.normal(0,Q)
        z=np.random.normal(x,R)
        K=P*(1/(R+P))
        x_hat=x_hat+K*(z-x_hat)        
        P=P-K*P
        
        Lx_hat.append(x_hat)
        Lz.append(z)
        Lt.append(t)
    return Lt, Lx_hat, Lz
```

**$$Q=10^{-5} :$$**

```python
def plot_Kalman(N, R=1, Q=1e-5, C=1)
    Lt, Lx_hat, Lz = Kalman(N, R=R, Q=Q, C=C)
    plt.figure(figsize=(15, 10))
    plt.plot(Lt, [C]*len(Lt), color='r', label="constante")
    plt.plot(Lt, Lx_hat, color='g', label="estimation")
    plt.scatter(Lt, Lz, color='b', alpha=0.1,label= "mesures")
    plt.legend()
    plt.show()
```
