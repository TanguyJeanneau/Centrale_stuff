# Filtre de Kalman Etendu :

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm
```

## 1) Pour la simulation :

$a_k=5$, amplitude de la sinusoïde

$\phi_k=0$, valeur de la phase de la sinusoïde

$\nu_0$, valeur de la fréquence de la sinusoïde

$\nu_e$=193.28 Hz, fréquence d'échantillonnage

Matrice de covariance de bruit d'état : 
$$Q=\begin{pmatrix}
2 10^{-5} &0 \\
0 &2 10^{-1}
\end{pmatrix}$$

Matrice de covariance du bruit de mesure: 

$$R=3$$

Matrice décrivant la dynamique du système: 

$$A=\begin{pmatrix}
1.0 &0 \\
0 &1.0
\end{pmatrix}$$


```python
Amplitude = 5
Phi = 0
Nu_e = 193.28
Nu_0 = 12
Te = 1 / Nu_e
nbEchantillon = 20000
t = np.arange(0, nbEchantillon , 1) * Te
```

```python
A = np.array([[1.0, 0.0],[0.0, 1.0]])
Q = np.array([[2e-5, 0.0],[0.0, 2e-1]])
```

```python
X = np.zeros(shape = (2, nbEchantillon))
Xhat = np.transpose([Amplitude, Phi])
X[:,0] = Xhat

# On applique l'équation qui caractérise la dynamique du système
for i in range(1, nbEchantillon, 1):
    X[:,i] = np.dot(A, X[:,i-1]) #+ np.transpose(np.random.multivariate_normal(np.transpose([0, 0]), Q))
```

```python
def equationMesure(x, t, Nu):
    Amplitude = x[0,:]
    Phi = x[1,:]
    Resultat = Amplitude * np.sin(2 * np.pi * Nu * t + Phi )
    return Resultat
```

```python
R = 3
sd = np.sqrt(R)
Y = np.zeros(shape = nbEchantillon)
# Calcul de la sortie réelle du système
# on applique l'équation de sortie
Y_real = equationMesure(X, t, Nu_0)
# Bruit de mesure
Bm = norm.rvs(0, sd, nbEchantillon)
# Calcul de la sortie bruitée
# Mesures dont on dispose pour faire l'estimation
Y = Y_real + Bm
```

```python
limite = 500
fig=plt.figure(figsize=(15,6))
plt.plot(t[0:limite], Y_real[0:limite], 'r')
plt.title('Le signal réel')
plt.xlabel('t (s)')
plt.ylabel('Amplitude')
plt.grid()
plt.show()
```

```python
limite = 500
fig=plt.figure(figsize=(15,6))
plt.plot(t[0:limite], Y[0:limite], 'ro')
plt.title('Les mesures bruitées')
plt.xlabel('t (s)')
plt.ylabel('Amplitude')
plt.grid()
plt.show()
```

```python
limite = 500
fig=plt.figure(figsize=(15,6))
plt.plot(t[0:limite], Y_real[0:limite], 'r')
plt.plot(t[0:limite], Y[0:limite], 'bo')
plt.title('Le signal réel et les mesures bruitées')
plt.xlabel('t (s)')
plt.ylabel('Amplitude')
plt.grid()
plt.show()
```

## Mise en oeuvre du filtre :

```python
def Kalman(z=Y, R=R, Q=Q,Nu_0=Nu_0,Te=Te):
    #Initialization :
    Lx_hat=[]
    Lt=[]
    #On initialise X :
    X=np.array([0,0])
    #signal z 
    N=len(z)
    #P :
    P=np.array([[0, 0.1],[0, 0.1]])
    #compteur k :
    k=0
    while k<N:
        #vk bruit d'état ->np.random.normal(0,Q)
        
        X=X+sum(np.random.normal(0,Q))
        
        P=P+Q
        
        k+=1
        
        #formule pour H :
        
        H=np.array([np.sin(2*np.pi*Nu_0*k*Te + X[1])*X[0]*np.cos(2*np.pi*Nu_0*k*Te+X[1])])
        
        S=H*P*H+R
        
        K=(P*H).dot(np.linalg.inv(S))
        
        #Formule pour h :
        
        h=X[0]*np.sin(2*np.pi*Nu_0*k*Te+X[1])
        
        epsilon=z[k]-h
        
        X=X+K*epsilon
        
        P=P-(K*H).dot(P)
        
        Lx_hat.append(X)
        Lt.append(t)
    return Lt, Lx_hat
```

```python
def plot_Kalman(z=Y, R=R, Q=Q,Nu_0=Nu_0,Te=Te):
    Lt, Lx_hat = Kalman(z=Y, R=R, Q=Q,Nu_0=Nu_0,Te=Te)
    plt.figure(figsize=(15, 10))
    plt.plot(Lt, Lx_hat, color='g', label="estimation")
    plt.legend()
    plt.show()
```

```python
plot_Kalman(z=Y, R=R, Q=Q,Nu_0=Nu_0,Te=Te)
```
