 ## Exercise 1: Ehrenfest model
 Consider a system of $K = 50$ particles (labeled from 1 to $K$) evolving in a closed box. The box is divided into two compartments in contact with each other, respectively identified by an index, 0 and 1. A hole at the interface between the two compartments allows the particles to move from one compartment to the other.

 Particle motion is modeled as follows: at each discrete time intant $n$, one particle is chosen uniformly at random and moved from its current compartment to the other. Let $X(n)$ denote the number of particles in compartment 0 at time $n$.

1\. Briefly justify that $\bigl(X(n) \bigr)_{n \in \mathbb{N}}$ is a Markov chain.

**$\bigl(X(n) \bigr)_{n \in \mathbb{N}}$ is a Markov chain because it is a family $\bigl(X(n) \bigr)_{n \in \mathbb{N}}$ of random variables  indexed by natural numbers and satisfying the markov property :**

$$ P(X(n)=i_n|X(n-1)=i_{n-1},..., X(0)=i_0)=P(X(n)=i_n|X(n-1)=i_{n-1})$$

2\. Recall the structure of the transition matrix, and encode it in Python (without any for loop).

 > Hint: use the function `numpy.diag`.

**The structure of the transition matrix is a matrix of $p_{i,j}$ the probability of getting from state $i$ to state $j$ at a given discrete time n.**

In this particular case, we can only go from one state $i$ to the state $i-1$ or $i+1$. 
At a certain time, the probability of getting from state $i$ to state $i+1$ is $p$ and the probability of getting from state $i$ to state $i-1$ is $1-p$. 

Therefore, the transition matrix is null, except the sub-diagonal and the upper diagonal, equal respectively to $1-p$ and $p$.

Since the particle moving from one compartiment to the other is chosen uniformly at random, $p$ at the given time $n$ is equal to $\frac{50-X(n)}{50}$.

We plot below the transition matrix for K=50 particles as a heatmap for more convenience : 

```python
def create_P(K):
    """
    Function create_P : to compute the transition matrix P associated with the Ehrenfest model
    @Inputs : 
        K : int, the number of states considered in the markov chain 
    @Outputs : 
        the transition matrix associated with the Ehrenfest model 
    """
    upper_diag = [(K-n)/K for n in range (0,K)]
    lower_diag =[n/K for n in range (1,K+1)]
    #Transition matrix P :
    return np.diag(upper_diag, k=1) + np.diag(lower_diag, k=-1)
#K=50 particles
K=50
#We create P :
P=create_P(K)
#Initialization of the matplotlib figure :
fig=plt.figure(figsize=(10,10))
#Plotting the matrix as a heatmap :
plt.title('Transition matrix with K=50 particles',size=15)
sns.heatmap(P)
```

3\. Numerically verify that the binomial distribution $\mathcal{B} (K, 1/2)$ is invariant for the chain. This invariant distribution will be denoted by $\pi$ in the rest of the exercise.

 > Hint: use `scipy.special.binom`.

**We write a function to create pi :** 

```python
def create_pi(K):
    """
    Function create_pi : to cumpute the invariant distribution pi for the Ehrenfest model
    @Inputs : 
        K : int, the number of states considered in the markov chain 
    @Outputs : 
        array, the invariant distribution pi computed from the binomial distribution B(K,1/2)
    """
    return np.array([scipy.special.binom(K,i)*(0.5)**i*(0.5)**(K-i) for i in range(K+1)])
#we compute pi with K=50 states :
pi=create_pi(50)
```

```python
diff = np.linalg.norm(np.dot(pi, P) - pi)
print('the difference in norm between pi*P and pi is : ', diff)
```

**We see that the difference between $\pi P$ and $\pi$ is very close to zero, so $\pi$ is indeed the invariant distribution for this chain.**

4\. Implement a Python function `ehrenfest` simulating a trajectory of the chain for a system of N particles, of initial distribution $\mu$. The maximum number of time steps will be controlled by an input parameter $n_\max$.

 > Remarks:
 > - for conveninence, you may define the state of a particle as a boolean variable. Changing the state of a particle will thus reduce to toggling the state of the associated boolean variable;
 > - note that, for any $i \in \{ 1, \dotsc , K \}$, imposing $X(0) = i$ means that the initial probability distribution $\mu$ is a Dirac centered in $i$.

```python
def ehrenfest(N, mu, nmax):
    """
    function ehrenfest(N, mu, nmax) simulating a trajectory of the chain for a system of N particles 
    @Inputs :
        - N, int : the number of particles
        - mu, array : the initial distribution
        - nmax, int : the maximum number of time steps
    @Outputs :
        -states, array : the successive states visited by the chain
    """
    P=create_P(N)
    next_step=np.random.choice(N+1, p=mu)
    states=[next_step]
    for i in range(nmax-1):
        next_step=np.random.choice(N+1, p=P[next_step])
        states.append(next_step)
    return states
```

 9.  a) Modify the function defined in 1. so that it returns the return time to state 0, defined as $T_{0,0} = \inf \bigl\{ n > 0, X(n) = 0 \mid X(0) = 0 \bigr\}$.

**We modify the function `ehrenfest(N, mu, nmax)` to obtain `ehrenfest_return_time(N,mu)` which returns the return time to state 0 :**

```python
def ehrenfest_return_time(N, mu):
    """
    function ehrenfest_return_time(N, mu) simulating the return time of the chain to state 0 
    @Inputs :
        - N, int : the number of particles
        - mu, array : the initial distribution
    @Outputs :
        -return_time, int : the return time to 0
    """
    P=create_P(N)
    next_step=np.random.choice(N+1, p=mu)
    return_time=0
    step=1
    while step != 0:
        next_step=np.random.choice(N+1, p=P[next_step])
        return_time+=1
        step=next_step
    return return_time
```

**We compute the return time for K=10 particles because it becomes difficult to compute with a large K** 

```python
#Number of particles : 
K=10
#We define mu so that the markov chain starts from state 0:
mu=np.array([1]+[0]*K)

print('return time :',  ehrenfest_return_time(K, mu))
```

 ## Exercise 2: Simulation of a discrete time homogeneous Markov chain.
 Let $\bigl( X(n) \bigr)_{n \in \mathbb{N}}$ be a discrete time homogeneous Markov chain defined by the following initial distribution $\mu$ and transition matrix $P$

\begin{equation}
     \mu = [0, 1, 0, 0, 0, 0], 
     %
     \quad
     %
     P = \begin{pmatrix}
       1/2   &1/2 &0  &0   &0   &0   \\
	 	1/4 &0   &0  &1/4 &1/4 &1/4   \\
       1/2   &0   &0  &0   &0   &1/2 \\
       0   &1/2 &0  &0   &1/2 &0   \\
       0   &1/3 &0  &1/3 &0   &1/3 \\
       0   &1/3 &1/3  &0 &1/3   &0
     \end{pmatrix}.
 \end{equation}

1\. Write a function `simulate_dthmc` simulating the trajectory of the Markov chain $\bigl( X(n) \bigr)_{n \in \mathbb{N}}$ for $n_{\max}$ time steps. The signature of the function should include the following elements:
   - list of inputs: transition matrix $P$, initial distribution $\mu$, number of time steps $n_{\max}$;
   - output: array of lenght $n_{\max}$ representing the trajectory.
   

```python
def simulate_dthmc(P,mu,nmax):
    """
    function simulate_dthmc(P, mu, nmax) simulating a trajectory of the chain for a system of N particles 
    @Inputs :
        - P, array : the transition matrix
        - mu, array : the initial distribution
        - nmax, int : the maximum number of time steps
    @Outputs :
        -states, array of length nmax : the successive states visited by the chain
    """
    next_step=np.random.choice(P.shape[0], p=mu)
    trajectory=[next_step]
    for i in range(nmax-1):
        next_step=np.random.choice(P.shape[0], p=P[next_step])
        trajectory.append(next_step)
    return np.array(trajectory)
```

2\. Simulate a trajectory of the chain for $n_{\max} = 2000$ starting from $X(0) = 1$. Plot the histogram of the states visited by the chain.
 

```python
#Initialization of P : 
P = np.array([
    [1/2  , 1/2  , 0  , 0  , 0  , 0],
    [1/4, 0  , 0  , 1/4, 1/4, 1/4],
    [1/2  , 0  , 0  , 0  , 0  , 1/2],
    [0  , 1/2, 0  , 0  , 1/2, 0  ], 
    [0  , 1/3, 0  , 1/3, 0  , 1/3],
    [0  , 1/3, 1/3, 0  , 1/3, 0  ]
])
#initialization of mu so that X(0)=1 :
mu=np.array([0,1,0,0,0,0])
#initialization of nmax :
nmax=2000
```

```python
#Computing the trajectory : 
trajectory=simulate_dthmc(P,mu,nmax)
#Plotting the histogram :
fig, ax = plt.subplots()
#Trick to have the bars of the histogram in front of each number of state associated : 
bins = np.arange(0, trajectory.max() + 1.5) - 0.5
_=ax.hist(trajectory, bins)
ax.set_xticks(bins+0.5)
plt.title('Histogram of the states visited by the chain :')
plt.xlabel('states')
plt.ylabel('Number of times each state is visited')
plt.show()
```

3\. Determine numerically the invariant distribution $\boldsymbol{\pi}$ of the chain (*e.g.*, based on an eigendecomposition [numpy.linalg.eig](https://docs.scipy.org/doc/numpy/reference/generated/numpy.linalg.eig.html)). Compare it to the histogram obtained in 2.
 

```python
# we compute the eigenvalues and the eigenvectors of P.T
eigenvalues, eigenvectors = np.linalg.eig(P.transpose())

#We search the eigenvector associated to the eigenvalue 1 :
pos_values_dom = np.argmax(eigenvalues)
vec_dom = eigenvectors[:, pos_values_dom]

# We verify that the eigenvalue is indeed equal to one 
assert abs(np.max(eigenvalues) - 1) < 1e-10

#We find the eigenvector associated and we make sure that the sum of its composants is equal to one :
pi = eigenvectors[:,pos_values_dom]/np.sum(vec_dom)
```

```python
#we compute the difference between pi*P and pi : 
diff = np.linalg.norm(np.dot(pi,P)-pi)
print('The difference in norm between pi*P and pi is : ', diff)
print('The sum of the composants of pi is : ', pi.sum())
```

Just like before, we have indeed numerically: 
    $$|\pi P - \pi|<10^{-15}|$$ 

 
 Thus,  we can say that $$\pi P = \pi$$ 
 
 Also, we verified that the sum of the composants of $\pi$ is equal to one.

```python
#Computing the trajectory : 
trajectory=simulate_dthmc(P,mu,nmax)
#Plotting the histogram :
fig, ax = plt.subplots()
#Trick to have the bars of the histogram in front of each number of state associated : 
bins = np.arange(0, trajectory.max() + 1.5) - 0.5
_=ax.hist(trajectory, bins,normed=True, label='Empirical histogram of the trajectory')
ax.set_xticks(bins+0.5)
plt.title('Histogram of the states visited by the chain :')
plt.xlabel('states')
plt.ylabel('Number of times each state is visited')
plt.plot(range(6), pi, color='r', label='Theoretical limit distribution pi')
plt.legend()
plt.show()
```

5\. For each state $i \in \{1, \dotsc, 5 \}$, simulate 1000 trajectories starting from the state $i$ until the return time to $i$. For each state, compute the (empirical) average return time. Compare with its theoretical value.
 

```python
def simulate_dthmc_return_time(P,mu,i):
    next_step=np.random.choice(P.shape[0], p=mu)
    return_time=0
    if i !=0:
        step=i-1
    else:
        step=1
    while step != i:
        next_step=np.random.choice(P.shape[0], p=P[next_step])
        return_time+=1
        step=next_step
    return return_time
```

```python
#We want to simulate n_traj=1000 trajectories
n_traj=1000
#We save the empirical average return time in 'avg_return_time' :
avg_return_time=[]

for i in range(6):
    #We count each return time in 'return_time' :
    return_time=[]
    #we define mu so that the markov chain starts from state i : 
    mu=np.array([0]*(i)+[1]+[0]*(5-i))
    for j in range(n_traj):
        return_time.append(simulate_dthmc_return_time(P,mu,i))
    avg_return_time.append(np.mean(return_time))

print('empirical mean return time for each state i : ', avg_return_time)
```

**the theoretical value of the average return time $T_{i,i}$ is $\frac{1}{\pi_i}$**

```python
print('theoretical mean return time for each state i : ', 1/pi)
```

**We can compare those values -theoretical and empirical- for each state i :** 

```python
for i in range(6):
    print('state ', i, " :")
    print("{rate:.2f} % of difference".format(rate=np.linalg.norm(avg_return_time[i]-1/pi[i])/avg_return_time[i] *100))
```

**Therefore, we can say that the empirical results are really close to the theoretical results.**
